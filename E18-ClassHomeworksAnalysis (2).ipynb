# Exercise 18.3

Create a classifier to predict the sex of each student



```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(max_features=20000, ngram_range=(1,2))
```

```python
vect = pd.DataFrame(vect.fit_transform(df['Text']).toarray(), index=df.index,columns=vect.get_feature_names())
```

```python
vect["total_W"]=data_nf['totalwords']
vect["sentXword"]=data_nf['sentXword'] 
vect["puntXword"]=data_nf['puntXword']
vect
```

```python
y = data_nf[["Sexo"]]
y = pd.DataFrame(y['Sexo'].map({'H':1,'M':0}),index = vect.index, columns=["Sexo"])
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=400, random_state=1, n_jobs=-1, bootstrap=True, max_depth=30, max_features='sqrt',
                            min_samples_leaf=1,min_samples_split=5)
```

```python
cross_val_score(clf, vect, y, cv=5, scoring='accuracy').mean()
```
